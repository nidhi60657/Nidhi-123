```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df = pd.read_csv('kidney_disease.csv')
```

```python
df
```

```python
to_drop = ['al','su','rbc','pc','pcc','ba','bgr','pcv','sod','pot','bu','wc','rc','htn','pe','ane']
df.drop(to_drop, inplace=True, axis=1)
```

```python
df
```

```python
df[:] = np.nan_to_num(df)
```

```python
df = df.set_index('id')
df
```

```python
X = df.iloc[: , 1:8]
```

```python
print(X)
```

```python
y = df.iloc[: , 8:9]
```

```python
 print(y)
```

```python
X.shape
```

```python
diabetes=pd.get_dummies(X['dm'],drop_first=True)
caad=pd.get_dummies(X['cad'],drop_first=True)
appete=pd.get_dummies(X['appet'],drop_first=True)
```

```python
X=pd.concat([X,diabetes,caad,appete],axis=1)
```

```python
X=X.drop(['dm','cad','appet'],axis=1)
```

```python
print(X)
```

```python
classcify=pd.get_dummies(y['classification'],drop_first=True)
y=y.drop(['classification'],axis=1)
y=pd.concat([y,classcify],axis=1)
#from sklearn.preprocessing import LabelEncoder
#le = LabelEncoder()
#y = le.fit_transform(y)
```

```python
print(y)
```

```python
y1=y.iloc[:,:-1]
```

```python
y1
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test ,y1_train, y1_test = train_test_split(X, y1, test_size = 0.2, random_state = 1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train  = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
import tensorflow as tf
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
classifier=Sequential()
```

```python
classifier.add(Dense(activation="relu",input_dim=11,units=6,kernel_initializer="uniform"))
```

```python
classifier.add(Dense(activation="relu",units=6,kernel_initializer="uniform"))
```

```python
classifier.add(Dense(activation="sigmoid",units=1,kernel_initializer="uniform"))
```

```python
classifier.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
classifier.fit(X_train ,y1_train, batch_size=7 ,epochs=20)
```

```python

y_pred = classifier.predict(X_test)
y_pred = (y_pred > 0.5)
```

```python
classifier.summary()
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score

cm=confusion_matrix(y1_test,y_pred)
```

```python
accuracy=accuracy_score(y1_test,y_pred)
```
